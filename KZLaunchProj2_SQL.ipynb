# Importing Section

```python
dfsection = pd.read_csv('g3dbsection.csv')
dfsection
```

```python
table_name = 'Section'
dfsection.to_sql(name=table_name, con=connection, if_exists='replace', index=False)
```

```python
query1 = """ SELECT count(course_id), semester, year_offered from Section
GROUP BY semester, year_offered
LIMIT 5 OFFSET 1"""
```

```python
result = pd.read_sql(query1, con=connection)
```

```python
result
```

```python
#most popular course
query2 = """ SELECT course_id, sum(num_registered), sum(max_capacity), 
sum(num_registered)/sum(max_capacity)*100 AS 'Percent Filled' from Section
GROUP BY course_id
ORDER BY sum(num_registered)/sum(max_capacity)*100 desc
LIMIT 30 OFFSET 1"""
```

```python
result2 = pd.read_sql(query2, con=connection)
result2
```

```python
#most popular time slot
query3 = """ SELECT time_slot, count(time_slot) from Section
GROUP BY time_slot
ORDER BY count(time_slot) desc
LIMIT 5"""
```

```python
result3 = pd.read_sql(query3, con=connection)
result3
```

```python
#most popular course
query4 = """ SELECT course_id AS 'Course', avg(num_registered) AS 'Avg # of Students Registered Per Section', semester, year_offered from Section
GROUP BY course_id, semester, year_offered
ORDER BY avg(num_registered) desc
LIMIT 30 OFFSET 1"""
```

```python
result4 = pd.read_sql(query4, con=connection)
result4
```

```python
#most popular course
query5 = """ SELECT course_id AS 'Course', avg(num_registered) AS 'Avg # of Students Registered Per Section' from Section
GROUP BY course_id
ORDER BY avg(num_registered) desc
LIMIT 30 OFFSET 1"""
result5 = pd.read_sql(query5, con=connection)
result5
```

```python
query6 = """ SELECT building AS 'Building', count(course_id) AS '# Classes Held' from Section
GROUP BY building
HAVING building != 'varchar(25)'
ORDER BY count(building) desc"""
result6 = pd.read_sql(query6, con=connection)
result6
```
